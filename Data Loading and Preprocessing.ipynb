```python
# ==== Import Required Libraries ====

import os
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from torchvision import transforms
from tqdm import tqdm
```

# Clean text (basic preprocessing)

```python
# ==== Text Cleaning ====

import re

def clean_caption(text):
    """
    This function removes special characters, digits
    and converts the text to clear lowercase caption.
    """
    text = text.lower()
    text = re.sub(r"[^a-zA-Z ]", "", text)  # remove special characters & digits
    text = re.sub(r"\s+", " ", text).strip()  # remove extra spaces
    return text

captions_df["clean_caption"] = captions_df["caption"].apply(clean_caption)

captions_df.head()
```

## Train / Validation / Test Split

```python
# ==== Train / Validation / Test Split ====

train_df, test_df = train_test_split(captions_df, test_size=0.20, random_state=42)
val_df, test_df = train_test_split(test_df, test_size=0.50, random_state=42)

print("Train size:", len(train_df))
print("Validation size:", len(val_df))
print("Test size:", len(test_df))
```

# Image Transformations (Resizing + Normalization)

```python
# ==== Image Transformations ====

image_transform = transforms.Compose([
    transforms.Resize((224, 224)),  # CLIP & BLIP expected size
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.5, 0.5, 0.5],
        std=[0.5, 0.5, 0.5]
    )
])
```

## Data Augmentation (Optional but improves accuracy)

```python
# ==== Data Augmentation for Training ====

augment_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.5, 0.5, 0.5],
        std=[0.5, 0.5, 0.5]
    )
])
```

# Function to load image safely

```python
# ==== Helper: Load Image Safely ====

def load_image(image_id):
    """
    Load an image given its ID. Apply transforms.
    """
    image_path = os.path.join(IMAGE_DIR, image_id)
    
    if not os.path.exists(image_path):
        print("Missing image:", image_id)
        return None
    
    img = Image.open(image_path).convert("RGB")
    return img
```

```python
# ==== Preview sample images with captions ====

def show_sample(df, n=3):
    sample = df.sample(n)
    
    for idx, row in sample.iterrows():
        img = load_image(row["image"])
        
        plt.figure(figsize=(4, 4))
        plt.imshow(img)
        plt.axis("off")
        plt.title(row["clean_caption"])
        plt.show()

# Show 3 training samples
show_sample(train_df, n=3)
```

# Create a PyTorch-compatible Dataset Class

```python
# ==== Custom Dataset Class for PyTorch ====

from torch.utils.data import Dataset

class ImageCaptionDataset(Dataset):
    def __init__(self, df, image_dir, transform=None):
        self.df = df
        self.image_dir = image_dir
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image_id = row["image"]
        caption = row["clean_caption"]
        
        img = load_image(image_id)
        if self.transform:
            img = self.transform(img)
        
        return img, caption
```

```python
# Cell 6 â€” BLIP Fine-Tuning Loop

num_epochs = 1 
MAX_STEPS = 5 # ðŸ’¡ FINAL CHANGE: Set MAX_STEPS to 5 for quickest completion

for epoch in range(num_epochs):
    blip_model.train()
    running_loss = 0.0

    # enumerate à¤•à¤¾ à¤‰à¤ªà¤¯à¥‹à¤— à¤•à¤°à¥‡à¤‚ à¤¤à¤¾à¤•à¤¿ à¤¹à¤® à¤¸à¥à¤Ÿà¥‡à¤ª à¤¨à¤‚à¤¬à¤° à¤œà¤¾à¤¨ à¤¸à¤•à¥‡à¤‚
    for step, (images, captions) in enumerate(tqdm(train_loader)):
        
        #  MAX_STEPS (5) à¤¤à¤• à¤ªà¤¹à¥à¤‚à¤š à¤—à¤ à¤¹à¥ˆà¤‚
        if step >= MAX_STEPS:
            print(f"\nTerminating epoch after {MAX_STEPS} steps for quick completion.")
            break 
            
        # Denormalization Fix (from [-1, 1] to [0, 1])
        denormalized_images = (images + 1) / 2 
        
        # Preprocess
        inputs = blip_processor(
            images=denormalized_images, 
            text=captions,
            padding='max_length',
            truncation=True,
            return_tensors="pt",
            do_rescale=False # Rescale Fix
        ).to(device)

        optimizer.zero_grad()
        outputs = blip_model(**inputs, labels=inputs["input_ids"])
        loss = outputs.loss

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # The loss output will now divide the total running loss by 5 steps
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/MAX_STEPS:.4f}")
```

```python
##  Cell 1: Import Libraries and Load Models (FIXED)

import torch
from torch.utils.data import DataLoader
from transformers import BlipProcessor, BlipForConditionalGeneration
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score
from tqdm import tqdm
import nltk

#  FIX: Download NLTK resources directly to avoid exceptions and AttributeErrors
print("Downloading necessary NLTK resources...")
try:
    nltk.download('wordnet', quiet=True)
    nltk.download('punkt', quiet=True)
    print("NLTK resources downloaded/checked.")
except Exception as e:
    print(f"Could not download NLTK resources: {e}")
    

# --- Model Loading Starts Here ---
# GPU check
device = "cuda" if torch.cuda.is_available() else "cpu"
print("\nUsing device:", device)

# Load BLIP Processor and Model
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained(
    "Salesforce/blip-image-captioning-base"
).to(device)

# Load the fine-tuned weights from Notebook 2
# Note: Ensure "blip_finetuned_captioner.pth" exists after running Cell 7 in Notebook 2
try:
    # Use map_location to ensure it loads correctly regardless of the training device
    blip_model.load_state_dict(torch.load("blip_finetuned_captioner.pth", map_location=device))
    print("Fine-tuned BLIP model loaded successfully.")
except FileNotFoundError:
    print("Warning: Fine-tuned weights not found (blip_finetuned_captioner.pth missing). Using pre-trained weights.")

blip_model.eval()
```

```python
## Cell 2: Define Test DataLoader

# Assuming test_dataset is available from previous notebooks
batch_size = 8 # Evaluation batch size (8) 
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"Test loader successfully created with batch size: {batch_size}")
print(f"Test set evaluation is ready.")
```

```python
## Cell 3: Caption Generation and Evaluation Loop

def denormalize_image(images):
    # This crucial fix from Notebook 2 ensures images are in [0, 1] range
    return (images + 1) / 2

all_generated_captions = []
all_reference_captions = []

# FIX: Set MAX_TEST_BATCHES to 5 for quick evaluation
MAX_TEST_BATCHES = 5 
batch_count = 0

with torch.no_grad():
    for images, captions_batch in tqdm(test_loader, desc="Generating Captions"):
        if batch_count >= MAX_TEST_BATCHES:
            break
        
        # 1. Preprocessing (Denormalization and to_device)
        denormalized_images = denormalize_image(images)
        
        inputs = blip_processor(
            images=denormalized_images,
            return_tensors="pt",
            do_rescale=False # Important fix to avoid rescaling error
        ).to(device)

        # 2. Caption Generation
        generated_ids = blip_model.generate(**inputs, max_length=16, num_beams=5) 
        
        # 3. Decode Generated Captions
        generated_captions = blip_processor.batch_decode(
            generated_ids, skip_special_tokens=True
        )

        # 4. Store Results
        all_generated_captions.extend(generated_captions)
        all_reference_captions.extend(captions_batch)
        
        batch_count += 1

print(f"\nCaption generation complete for {len(all_generated_captions)} samples.")
```

```python
##  Cell 4: Calculate BLEU and METEOR Scores

bleu_scores = []
meteor_scores = []
smooth = SmoothingFunction() 

for candidate, references in zip(all_generated_captions, all_reference_captions):
    
    # Candidate should be a list of words (tokenized)
    candidate_tokens = candidate.split()
    
    # References should be a list of lists of words (tokenized references)
    reference_tokens = [ref.split() for ref in references]
    
    # 1. Calculate BLEU-4 Score (uses tokenized inputs)
    bleu = sentence_bleu(
        reference_tokens, candidate_tokens, 
        weights=(0.25, 0.25, 0.25, 0.25), 
        smoothing_function=smooth.method4
    )
    bleu_scores.append(bleu)

    # 2. Calculate METEOR Score (FIXED to use tokenized inputs)
    # METEOR needs pre-tokenized inputs (list of words for both reference and candidate).
    meteor = meteor_score(reference_tokens, candidate_tokens) 
    meteor_scores.append(meteor)

# Calculate averages
avg_bleu = sum(bleu_scores) / len(bleu_scores) if bleu_scores else 0
avg_meteor = sum(meteor_scores) / len(meteor_scores) if meteor_scores else 0

print("--- Captioning Metrics ---")
print(f"Total Samples Evaluated: {len(all_generated_captions)}")
print(f"Average BLEU-4 Score: {avg_bleu:.4f}")
print(f"Average METEOR Score: {avg_meteor:.4f}")
print("--------------------------")
```

```python
# Note on Captioning Metrics:
# The significantly low BLEU and METEOR scores are expected because 
# the BLIP model was intentionally fine-tuned for only 5 steps/batches 
# (as shown in Notebook 2) for the sake of quick demonstration and 
# limited compute resources. This is insufficient training for a large model.
print("\n[NOTE: Low BLEU/METEOR scores are expected due to insufficient training (only 5 steps) performed for quick demonstration.]")
```

```python
## Cell 5: Image Retrieval Evaluation (Recall@K)

from transformers import CLIPProcessor, CLIPModel
import numpy as np

# Load CLIP Model (Assuming it was loaded in Cell 1, if not, load it here)
try:
    if 'clip_model' not in globals():
        clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
        clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
except Exception:
    pass # Assume model is loaded if this fails

clip_model.eval()
print("CLIP model is ready for evaluation.")

# --- Recall@K Calculation Function ---

def calculate_recall_at_k(all_image_features, all_text_features, k_values=[1, 5]):
    """Calculates Image-to-Text Recall@K (I -> T)."""
    
    # Calculate similarity matrix (dot product of image and text features)
    similarity_matrix = all_image_features @ all_text_features.T
    
    num_images = similarity_matrix.shape[0]
    num_references = 5 # Flickr8k has 5 captions per image

    # Find the indices of the best matching captions for each image
    sorted_indices = torch.argsort(similarity_matrix, descending=True, dim=1)
    
    recall_results = {}
    
    for k in k_values:
        correct_retrievals = 0
        
        for img_idx in range(num_images):
            # Top K retrieved indices for the current image
            top_k_indices = sorted_indices[img_idx, :k]
            
            # Define the range of correct caption indices for this image
            start_idx = img_idx * num_references
            end_idx = start_idx + num_references
            correct_range = set(range(start_idx, end_idx))
            
            # Check if any index in top_k_indices falls within the correct_range
            retrieved_set = set(top_k_indices.cpu().numpy())
            
            if len(retrieved_set.intersection(correct_range)) > 0:
                correct_retrievals += 1
                
        recall = correct_retrievals / num_images
        recall_results[f"R@{k}"] = recall

    return recall_results

# --- Data Preparation for Retrieval ---

# Use the same subset size (40 samples) as Cell 3 for consistency
MAX_TEST_SAMPLES = len(all_generated_captions) 
num_samples = 0
all_image_features = []
all_text_features = []

with torch.no_grad():
    for images, captions_batch in tqdm(test_loader, desc="Extracting Features"):
        if num_samples >= MAX_TEST_SAMPLES:
            break
            
        # Adjust batch size for the last batch if needed
        batch_size = images.size(0)
        if num_samples + batch_size > MAX_TEST_SAMPLES:
             images = images[:MAX_TEST_SAMPLES - num_samples]
             captions_batch = captions_batch[:MAX_TEST_SAMPLES - num_samples]
             batch_size = images.size(0)

        # 1. Denormalize Image (Crucial fix)
        denormalized_images = (images + 1) / 2 
        
        # 2. Process Image and get Features
        image_inputs = clip_processor(images=denormalized_images, return_tensors="pt").to(device)
        image_features = clip_model.get_image_features(**image_inputs)
        image_features = image_features / image_features.norm(p=2, dim=-1, keepdim=True) # Normalize
        all_image_features.append(image_features)

        # 3. Process ALL 5 Captions per image and get Features
        # Flatten the list of captions (5 captions * batch_size)
        flat_captions = [cap for sublist in captions_batch for cap in sublist]
        text_inputs = clip_processor(text=flat_captions, return_tensors="pt", padding=True).to(device)
        text_features = clip_model.get_text_features(**text_inputs)
        text_features = text_features / text_features.norm(p=2, dim=-1, keepdim=True) # Normalize
        all_text_features.append(text_features)
        
        num_samples += batch_size

# Concatenate all features
all_image_features = torch.cat(all_image_features, dim=0)
all_text_features = torch.cat(all_text_features, dim=0)

# 4. Calculate Recall@K
recall_scores = calculate_recall_at_k(all_image_features, all_text_features, k_values=[1, 5])

print("\n--- Image Retrieval Metrics (Image to Text) ---")
print(f"Total Images Evaluated: {all_image_features.shape[0]}")
for k, score in recall_scores.items():
    print(f"Recall@{k}: {score:.4f}")
print("---------------------------------------------")
```

```python
# Note on Image Retrieval Metrics:
# The Recall@K scores are extremely low (close to 0) because the retrieval 
# model (CLIP) was used without fine-tuning and the evaluation was performed 
# on a very small subset of the data (40 samples). 
# A full retrieval system requires pre-calculating and indexing features 
# for the entire dataset for meaningful results.
print("\n[NOTE: Low Retrieval scores (Recall@K) are also expected as the CLIP model was not fine-tuned and the test set was limited to 40 samples, which is insufficient for meaningful retrieval evaluation.]")
```

```python
## Cell 1: Pre-calculate and Save CLIP Features for Fast Retrieval

import torch
from torch.utils.data import Subset # Import Subset for safe slicing
from transformers import CLIPProcessor, CLIPModel
from tqdm import tqdm
import numpy as np

# --- Configuration ---
NUM_DEMO_SAMPLES = 500  # Number of samples to include in the demo index
FEATURE_FILE_NAME = 'test_features.pt'

# --- Load Models & Setup Device ---
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Load CLIP Model
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_model.eval()

# --- Data Preparation (Fix for slicing error) ---
# FIX: Use Subset to safely slice the test_dataset without causing the Pandas 'Series' error.
if len(test_dataset) > NUM_DEMO_SAMPLES:
    demo_dataset = Subset(test_dataset, range(NUM_DEMO_SAMPLES))
    print(f"Created demo_dataset with {len(demo_dataset)} samples using Subset.")
else:
    demo_dataset = test_dataset
    print(f"Using full test_dataset, containing {len(demo_dataset)} samples.")

# --- Feature Extraction Loop ---
all_image_features = []
all_text_features = []
all_demo_images = []
all_demo_texts = []

with torch.no_grad():
    for image, captions in tqdm(demo_dataset, desc="Pre-calculating CLIP Features"):
        
        # 1. Denormalize Image (Crucial fix from previous notebooks, converts [-1, 1] to [0, 1])
        denormalized_image = (image + 1) / 2
        
        # 2. Process Image and get Normalized Features
        image_inputs = clip_processor(images=denormalized_image, return_tensors="pt").to(device)
        image_feature = clip_model.get_image_features(**image_inputs)
        image_feature = image_feature / image_feature.norm(p=2, dim=-1, keepdim=True) # L2 Normalization
        all_image_features.append(image_feature.squeeze(0))
        
        # 3. Process the First Caption (Captions[0]) and get Normalized Features
        # We only index one reference caption per image for the Text -> Image search demo.
        first_caption = captions[0]
        text_inputs = clip_processor(text=first_caption, return_tensors="pt", padding=True).to(device)
        text_feature = clip_model.get_text_features(**text_inputs)
        text_feature = text_feature / text_feature.norm(p=2, dim=-1, keepdim=True) # L2 Normalization
        all_text_features.append(text_feature.squeeze(0))

        # 4. Store Raw Data for Display in Streamlit
        all_demo_images.append(image)
        all_demo_texts.append(first_caption)

# --- Save Features ---
image_features_tensor = torch.stack(all_image_features).cpu()
text_features_tensor = torch.stack(all_text_features).cpu()

torch.save({
    'image_features': image_features_tensor,
    'text_features': text_features_tensor,
    'images': all_demo_images,
    'captions': all_demo_texts,
}, FEATURE_FILE_NAME)

print(f"\nSaved CLIP features for {image_features_tensor.shape[0]} samples to {FEATURE_FILE_NAME}")
```

```python
# app.py content - Save this code to a file named 'app.py'

import streamlit as st
import torch
from transformers import BlipProcessor, BlipForConditionalGeneration
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import numpy as np
import io

# --- 1. CONFIGURATION ---
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
FEATURE_PATH = 'test_features.pt'
WEIGHTS_PATH = 'blip_finetuned_captioner.pth'
TOP_K_RETRIEVAL = 3

# --- 2. MODEL LOADING (Caching for efficiency) ---

@st.cache_resource
def load_models():
    # BLIP Model (Captioning)
    blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
    blip_model = BlipForConditionalGeneration.from_pretrained(
        "Salesforce/blip-image-captioning-base"
    ).to(DEVICE)
    try:
        # Load fine-tuned weights
        blip_model.load_state_dict(torch.load(WEIGHTS_PATH, map_location=DEVICE))
    except FileNotFoundError:
        st.warning(f"BLIP fine-tuned weights ({WEIGHTS_PATH}) not found. Using base model.")
    blip_model.eval()

    # CLIP Model (Retrieval)
    clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
    clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(DEVICE)
    clip_model.eval()

    # Pre-calculated Features
    try:
        features_data = torch.load(FEATURE_PATH)
        st.success(f"Loaded {len(features_data['image_features'])} features for retrieval.")
        return blip_processor, blip_model, clip_processor, clip_model, features_data
    except FileNotFoundError:
        st.error(f"Feature file '{FEATURE_PATH}' not found. Retrieval demo disabled.")
        return blip_processor, blip_model, clip_processor, clip_model, None

blip_processor, blip_model, clip_processor, clip_model, features_data = load_models()

# --- 3. HELPER FUNCTIONS ---

def get_caption(pil_image):
    # Generate the caption
    inputs = blip_processor(images=pil_image, return_tensors="pt").to(DEVICE)
    out = blip_model.generate(**inputs, max_length=16, num_beams=5)
    return blip_processor.decode(out[0], skip_special_tokens=True)

def retrieve_images(query_text, k=TOP_K_RETRIEVAL):
    if features_data is None: return [], []
    
    # Get Text Feature
    text_inputs = clip_processor(text=query_text, return_tensors="pt", padding=True).to(DEVICE)
    text_feature = clip_model.get_text_features(**text_inputs)
    text_feature = text_feature / text_feature.norm(p=2, dim=-1, keepdim=True)

    # Calculate Similarity with Saved Image Features
    img_features = features_data['image_features'].to(DEVICE)
    similarity = (text_feature @ img_features.T).squeeze(0)
    
    # Get Top K indices
    top_k_indices = torch.argsort(similarity, descending=True)[:k].cpu().numpy()
    
    # Fetch results
    retrieved_images = [features_data['images'][i] for i in top_k_indices]
    captions = [features_data['captions'][i] for i in top_k_indices]
    
    return retrieved_images, captions

# --- 4. STREAMLIT UI ---

st.set_page_config(layout="wide", page_title="BLIP & CLIP Demo")
st.title(" Visual-Language Model Demo")
st.markdown("---")

# Sidebar for controls
st.sidebar.header("Controls")
demo_mode = st.sidebar.radio("Select Demo Mode:", ["Image Captioning", "Semantic Search (Text to Image)"])

if demo_mode == "Image Captioning":
    st.header("1. Image Captioning (BLIP)")
    
    uploaded_file = st.file_uploader("Upload an Image to Caption", type=["jpg", "jpeg", "png"])
    
    if uploaded_file is not None:
        pil_image = Image.open(uploaded_file).convert("RGB")
        
        col1, col2 = st.columns(2)
        
        with col1:
            st.image(pil_image, caption='Uploaded Image', use_column_width=True)
            
        with col2:
            with st.spinner('Generating caption...'):
                caption = get_caption(pil_image)
            st.subheader("Generated Caption:")
            st.info(caption)

elif demo_mode == "Semantic Search (Text to Image)":
    st.header("2. Semantic Search (CLIP)")
    
    query_text = st.text_input(
        "Enter a description to search for similar images:",
        "a dog jumping in the grass"
    )
    
    if query_text and features_data:
        st.markdown(f"Searching for **'{query_text}'**...")
        
        #  Ensure the f-string is properly terminated here
        with st.spinner(f'Retrieving top {TOP_K_RETRIEVAL} similar images...'): 
            retrieved_images_tensor, captions = retrieve_images(query_text)
            
        st.subheader(f"Top {TOP_K_RETRIEVAL} Retrieved Images:")
        
        cols = st.columns(TOP_K_RETRIEVAL)
        
        for i, (img_tensor, cap) in enumerate(zip(retrieved_images_tensor, captions)):
            # Convert tensor back to PIL image for display
            # Tensors are [C, H, W] in [-1, 1] range. Denormalize to [0, 1] then convert to PIL
            img_tensor_denorm = (img_tensor + 1) / 2
            img_array = img_tensor_denorm.permute(1, 2, 0).cpu().numpy() # C H W -> H W C
            img_pil = Image.fromarray((img_array * 255).astype(np.uint8))
            
            with cols[i]:
                st.image(img_pil, caption=f"Caption: {cap}", use_column_width=True)

    elif features_data is None:
        st.error("Cannot run Semantic Search. The 'test_features.pt' file is missing or failed to load.")

st.markdown("---")
st.caption("Models: BLIP (Finetuned for Captioning) and CLIP (Pre-trained for Search).")
```
